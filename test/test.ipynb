```python
# Understanding the biomass function
import numpy as np
Ro = 12.0 #[g/cm2]
k1 = 0.1
du = 0.0
db = 30.0
top = Ro * pow(np.e, -k1*du)
bottom = Ro * pow(np.e, -k1*db)
print(top)
print(bottom)
print(bottom-top / -k1)
print(Ro * (pow(np.e, -k1*bottom) - pow(np.e, -k1*top) / -k1))
```

```python
from matplotlib import pyplot as plt
xs1 = [x for x in range(0, 31)]
ys1 = [Ro * pow(np.e, -k1*x) for x in xs1]
R_10 = Ro * pow(np.e, -k1*10)
print(R_10)
xs2 = [x for x in range(0, 21)]
ys2 = [R_10 * pow(np.e, -k1*x) for x in xs2]
plt.plot(xs1, ys1)
xs2plot = [x+10 for x in xs2]
plt.plot(xs2plot, ys2)
```

```python
def update(**kwargs):
    a = kwargs['a']
    b = kwargs['b']
    return called_by_update(a=a, b=b)

def called_by_update(a: float, b: float):
    return a + b

def call_update(a: float, b: float):
    return update(a=a, b=b)

c = call_update(3, 4)
print(c)
```
